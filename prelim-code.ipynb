```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Loading up

```python
train_features_csv = "./data/dengue_features_train.csv"
train_labels_csv = "./data/dengue_labels_train.csv"
```

```python
features = pd.read_csv(train_features_csv)
labels = pd.read_csv(train_labels_csv)
```

```python
# adding the label to features and renaming to "train" for simplicity
features["total_cases"] = labels.total_cases
```
