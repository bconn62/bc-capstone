```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split, cross_val_score

from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import regularizers
from keras.callbacks import EarlyStopping
```

```python
# the pred_maker!

def pred_maker(test_set, model, csv_name, polyfunc=None, scalerfunc=None, city=None):
    
    if city is not None:
        if city == "sj":
            test_set = test_set[test_set["city"]=="sj"].copy()
#             test_set.drop("city", axis=1, inplace=True)
        if city == "iq":
            test_set = test_set[test_set["city"]=="iq"].copy()
#             test_set.drop("city", axis=1, inplace=True)

    # preliminary preparation of test set
    test_set.drop("week_start_date", axis=1, inplace=True)
    test_set.drop("Unnamed: 0", axis=1, inplace=True)
    
    # this is to pull from for later in the prediction dataframe
#     if city is None:
    city_df = test_set["city"].copy()
    year_df = test_set["year"].copy()
    weekofyear_df = test_set["weekofyear"].copy()
    
    #dummying cities
    test_set = pd.get_dummies(test_set)
    
    # prepping weather features
    test_non_weather = "year weekofyear ndvi_ne ndvi_nw ndvi_se ndvi_sw city_iq city_sj".split()
    test_weather_features = [x for x in test_set.columns if x not in test_non_weather]
    
    # if there is a polynomial features function
    if polyfunc is not None:
        # calling the func and making a dataframe
        pf_test = polyfunc.transform(test_set[test_weather_features])
        test_weather = pd.DataFrame(data=pf_test, columns=polyfunc.get_feature_names(test_weather_features))
        
        test_set.drop(labels=weather_features, axis=1, inplace=True)
        
        # fixing the indices
        test_set.reset_index(drop=True, inplace=True)
        test_weather.reset_index(drop=True, inplace=True)

        test_set['id'] = test_set.index
        test_weather['id'] = test_weather.index
        
        #merging dataframes
        test_set = test_set.merge(test_weather, how='left', on='id')
        test_set.drop("id", axis=1, inplace=True)
        
    # if there is a scaler function
    if scalerfunc is not None:
        test_set = scalerfunc.transform(test_set)
    
    # making predictions
    pred_array = model.predict(test_set)
    
    #setting up prediction dataframe
    pred_df = pd.DataFrame()
#     if city is None:
    pred_df["city"] = city_df
    pred_df["year"] = year_df
    pred_df["weekofyear"] = weekofyear_df
    pred_df["total_cases"] = pred_array
    
    pred_df = pred_df.round({"total_cases":0})
    
#     pred_df.to_csv("./submissions/"+csv_name, index=False)

    if city is None:
        #taking their sample submission...
        sample_csv = "./data/submission_format.csv"
        sample = pd.read_csv(sample_csv)
    
        #... and making it my own
        sample.total_cases = pred_array
        pred_df = sample.copy()
        pred_df["total_cases"] = pred_df["total_cases"].astype("int64")
    
        pred_df.to_csv("./submissions/"+csv_name, index=False)
    
        return pred_df
    
    return pred_df
```

### Load up

```python
train_sj_csv = "./data/train_sj.csv"
train_iq_csv = "./data/train_iq.csv"
```

```python
train_sj = pd.read_csv(train_sj_csv)
train_iq = pd.read_csv(train_iq_csv)
```

```python
train_sj.drop("Unnamed: 0", axis=1, inplace=True)
train_iq.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
train_sj.drop("week_start_date", axis=1, inplace=True)
train_iq.drop("week_start_date", axis=1, inplace=True)
```

```python
# train_sj.drop("city", axis=1, inplace=True)
# train_iq.drop("city", axis=1, inplace=True)
```

```python
train_sj = pd.get_dummies(train_sj)
train_iq = pd.get_dummies(train_iq)
```

```python
train_sj.head()
```

#### poly weather

```python
non_weather = "year weekofyear ndvi_ne ndvi_nw ndvi_se ndvi_sw total_cases city_sj city_iq".split()
weather_features = [x for x in train_sj.columns if x not in non_weather]
```

```python
pf_sj = PolynomialFeatures(degree=2, include_bias=False)
pf_iq = PolynomialFeatures(degree=2, include_bias=False)

pf_sj_feat = pd.DataFrame(data=pf_sj.fit_transform(train_sj[weather_features]), columns=pf_sj.get_feature_names(weather_features))
pf_iq_feat = pd.DataFrame(data=pf_iq.fit_transform(train_iq[weather_features]), columns=pf_iq.get_feature_names(weather_features))
```

```python
train_sj_weather = train_sj.copy()
train_sj_weather.drop(labels=weather_features, axis=1, inplace=True)
train_sj_weather = train_sj_weather.join(pf_sj_feat, how="outer")
```

```python
train_iq_weather = train_iq.copy()
train_iq_weather.drop(labels=weather_features, axis=1, inplace=True)
train_iq_weather = train_iq_weather.join(pf_iq_feat, how="outer")
```

### 1a. Poly weather, separate cities, SJ

```python
target = "total_cases"
sj_feats = [x for x in train_sj_weather.columns if x != target]
X_sj = train_sj_weather[sj_feats]
y_sj = train_sj_weather[target]
```

```python
Xsj_train, Xsj_test, ysj_train, ysj_test = train_test_split(X_sj,y_sj)
```

```python
ss_sj = StandardScaler()
Xsjs_train = ss_sj.fit_transform(Xsj_train)
Xsjs_test = ss_sj.transform(Xsj_test)
```

```python
network_sj = Sequential()

network_sj.add(Dense(158, activation="relu", input_dim=Xsjs_train.shape[1], kernel_regularizer=regularizers.l1(.01)))
network_sj.add(Dropout(.5))
network_sj.add(Dense(316, activation="relu", kernel_regularizer=regularizers.l1(.01)))
network_sj.add(Dropout(.5))
network_sj.add(Dense(79, activation="relu", kernel_regularizer=regularizers.l1(.01)))
network_sj.add(Dropout(.5))
network_sj.add(Dense(1, activation=None, kernel_regularizer=regularizers.l1(.01)))

network_sj.compile(loss="mean_absolute_error", optimizer="adam")

network_sj.fit(Xsjs_train, ysj_train, validation_data=(Xsjs_test, ysj_test), epochs=500)
```

### 1b. poly weather, separate cities, IQ

```python
target = "total_cases"
iq_feats = [x for x in train_iq_weather.columns if x != target]
X_iq = train_iq_weather[iq_feats]
y_iq = train_iq_weather[target]
```

```python
Xiq_train, Xiq_test, yiq_train, yiq_test = train_test_split(X_iq,y_iq)
```

```python
ss_iq = StandardScaler()
Xiqs_train = ss_iq.fit_transform(Xiq_train)
Xiqs_test = ss_iq.transform(Xiq_test)
```

```python
network_iq = Sequential()

network_iq.add(Dense(158, activation="relu", input_dim=Xiqs_train.shape[1], kernel_regularizer=regularizers.l1(.01)))
network_iq.add(Dropout(.5))
network_iq.add(Dense(316, activation="relu", kernel_regularizer=regularizers.l1(.01)))
network_iq.add(Dropout(.5))
network_iq.add(Dense(79, activation="relu", kernel_regularizer=regularizers.l1(.01)))
network_iq.add(Dropout(.5))
network_iq.add(Dense(1, activation=None, kernel_regularizer=regularizers.l1(.01)))

network_iq.compile(loss="mean_absolute_error", optimizer="adam")

network_iq.fit(Xiqs_train, yiq_train, validation_data=(Xiqs_test, yiq_test), epochs=500)
```

```python
test = pd.read_csv("./data/test.csv")
preds_sj = pred_maker(test, network_sj, "bleh", polyfunc=pf_sj, scalerfunc=ss_sj, city="sj")
```

```python
test = pd.read_csv("./data/test.csv")
preds_iq = pred_maker(test, network_iq, "bleh", polyfunc=pf_iq, scalerfunc=ss_iq, city="iq")
```

```python
merged_preds = pd.concat(objs=[preds_sj, preds_iq])
```

```python
sample_csv = "./data/submission_format.csv"
sample = pd.read_csv(sample_csv)
    
sample.total_cases = merged_preds["total_cases"]
temp_df = sample.copy()
temp_df["total_cases"] = temp_df["total_cases"].astype("int64")
    
temp_df.to_csv("./submissions/poly_separate_nn.csv", index=False)
```
