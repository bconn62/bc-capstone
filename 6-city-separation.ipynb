### Separating SJ from IQ

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_csv = "./data/cleaned_train.csv"
train = pd.read_csv(train_csv)
```

```python
train.head()
```

```python
train.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
sj = train["city"]=="sj"
iq = train["city"]=="iq"
```

```python
train_sj = train[sj]
train_iq = train[iq]
```

```python
train_sj.head()
```

```python
train_iq.head()
```

```python
train_sj.to_csv("./data/train_sj.csv")
train_iq.to_csv("./data/train_iq.csv")
```
