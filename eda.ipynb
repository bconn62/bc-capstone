```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_csv = "./data/train.csv"
```

```python
train = pd.read_csv(train_csv)
```

```python
avg_case_per_week = train.groupby("weekofyear").mean()["total_cases"]
plt.plot(avg_case_per_week)
```

```python
year_totals = train.groupby("year").sum()["total_cases"]
plt.plot(year_totals)
```
